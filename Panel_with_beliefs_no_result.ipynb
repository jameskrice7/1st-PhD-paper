```python
import pyreadstat

# Specify the path to your .sav file
file_path = "C:/Users/jr24440/Downloads/CCAM SPSS Data 2008-2023.sav"

# Read the .sav file
df_CCAM, meta = pyreadstat.read_sav(file_path)

# Display the first few rows of the dataframe
print(df_CCAM.head())
```

```python
# Ensure 'year' is numeric (in case of any non-numeric entries)
df_CCAM['year'] = pd.to_numeric(df_CCAM['year'], errors='coerce')

# Drop rows with missing or invalid year values
df_CCAM = df_CCAM.dropna(subset=['year'])

# Convert to integer
df_CCAM['year'] = df_CCAM['year'].astype(int)

# Map coded year values so that 1 → 2008, 2 → 2009, ..., etc.
df_CCAM['year'] = df_CCAM['year'].map(lambda x: 2007 + x)

# Confirm the transformation
print(df_CCAM['year'].unique())
```

```python
df_CCAM = df_CCAM.rename(columns={'year': 'Date'})
```

```python
# Step 1: Convert 'Date' to datetime if not already
full_panel['Date'] = pd.to_datetime(full_panel['Date'])

# Step 2: Create a new column 'Year' extracted from 'Date'
full_panel['Year'] = full_panel['Date'].dt.year

# Step 3: Group by 'Year' and aggregate numeric columns
df_yearly_agg = full_panel.groupby('Year').mean(numeric_only=True).reset_index()
```

```python
import statsmodels.api as sm
import pandas as pd

# Ensure 'Date' is treated as categorical for fixed effects
df_merged['Date'] = df_merged['Date'].astype(str)

# Define your dependent variable
y = df_merged['ESG Score']

# Define your independent variables including CCAM controls and clusters
X = df_merged[
    [
        "misinfo_count_normal", "total_tweets_misinfo", "misinfo_count_misinfo", "total_tweets_normal",
        "cluster_Business Impact", "cluster_Research", "Carbon Tax", "Climate Summits", "Extreme Temperatures", "cluster_Societal Debate", "Carbon Tax", "Cities", "Ecosystems", "Corporations/Investments",
        "Food Shortage/Poverty", "UMCSENT", "cpu_index", "GEPU_current",
        "worry", "harm_personally", "sci_consensus", "discuss_GW", "reg_utilities", "hear_GW_media", "fund_research", "harm_dev_countries", "harm_US", "cause_original"
    ]
]

# Add year dummies for fixed effects
X = pd.concat([X, pd.get_dummies(df_merged['Date'], prefix='year', drop_first=True)], axis=1)

# Step 5: Ensure all variables are numeric
X = X.apply(pd.to_numeric)
y = pd.to_numeric(y)

# Add constant
X = sm.add_constant(X)

# Fit the model
model = sm.OLS(y, X).fit(cov_type='HC3')  # HC3 for robust standard errors

# Show results
print(model.summary())
```

```python
import numpy as np
import statsmodels.api as sm

# STEP 1: Ensure y and X are pulled from df_merged
y = df_merged["Aggregate"]
X = df_merged[
    [
        "misinfo_count_normal", "total_tweets_misinfo", "misinfo_count_misinfo", "total_tweets_normal",
        "cluster_Business Impact", "cluster_Research", "Carbon Tax", "Climate Summits", "Extreme Temperatures", "cluster_Societal Debate", "Carbon Tax", "Cities", "Ecosystems", "Corporations/Investments",
        "Food Shortage/Poverty", "UMCSENT", "cpu_index", "GEPU_current",
        "worry", "harm_personally", "sci_consensus", "discuss_GW", "reg_utilities", "hear_GW_media", "fund_research", "harm_dev_countries", "harm_US", "cause_original"
    ]
]

# STEP 2: Convert all columns to numeric (force coercion)
X = X.apply(pd.to_numeric, errors='coerce')
y = pd.to_numeric(y, errors='coerce')

# STEP 3: Drop any rows with missing or non-numeric values
df_model = pd.concat([y, X], axis=1).dropna()

# STEP 4: Re-split the cleaned dataframe
y_clean = df_model.iloc[:, 0].astype(float)
X_clean = df_model.iloc[:, 1:].astype(float)

# STEP 5: Add constant after cleaning
X_clean = sm.add_constant(X_clean)

# STEP 6: Fit the OLS model with robust standard errors
model = sm.OLS(y_clean, X_clean).fit(cov_type='HC3')
print(model.summary())
```

```python
# Step 1: Ensure column names are clean
X_clean.columns = X_clean.columns.astype(str).str.strip().str.replace(r"[^\w]", "_", regex=True)

# Step 2: Ensure no duplicates
X_clean = X_clean.loc[:, ~X_clean.columns.duplicated()]

# Step 3: Same for y (though it's usually safe)
y_clean.name = "Aggregate"

# Step 4: Fit the model again
model = sm.OLS(y_clean, X_clean).fit(cov_type='HC3')
print(model.summary())
```
