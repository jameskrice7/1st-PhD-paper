```python
from transformers import pipeline

# Example: Initialize a hate speech classifier (ensure model suitability)
hate_classifier = pipeline("text-classification", model="unitary/toxic-bert")

# Example: Initialize a fake news/misinformation classifier if available
misinfo_classifier = pipeline("text-classification", model="harshhmaniya/fake-news-classification-distilbert-fine-tuned")

# Define a simple function to process a single text sample
def analyze_text(text):
    # Get hate speech predictions
    hate_result = hate_classifier(text, truncation=True)
    # Get misinformation predictions
    # Truncate the input text to the maximum sequence length of the model
    misinfo_result = misinfo_classifier(text[:512], truncation=True)

    # Depending on your task, extract scores/labels.
    # Here we assume each pipeline returns a list of dictionaries with 'label' and 'score'
    return {
        'hate': hate_result,
        'misinformation': misinfo_result
    }

# Process a sample from your dataset (assuming you have a split named 'train' and a 'text' field)
sample_article = ds['train'][0]['text']
analysis = analyze_text(sample_article)
print(analysis)
```

```python
# Randomly sample 100 articles (example)
sample_ds = ds['train'].shuffle(seed=42).select(range(100))

results = []
for article in sample_ds:
    text = article['text']
    title = article['title']  # Ensure your dataset has a "title" field
    analysis = analyze_text(text)

    # Each pipeline returns a list of dictionaries, so we extract the first entry
    hate = analysis['hate'][0]
    misinfo = analysis['misinformation'][0]

    results.append({
        'title': title,
        'hate_label': hate['label'],
        'hate_score': hate['score'],
        'misinfo_label': misinfo['label'],
        'misinfo_score': misinfo['score'],
    })

import pandas as pd
df_results = pd.DataFrame(results)
print(results)
```

```python
from datasets import load_dataset
from transformers import pipeline

# 2. Initialize zero-shot classification pipeline
classifier = pipeline(
    task="zero-shot-classification",
    model="facebook/bart-large-mnli",
    device=-1  # If GPU is available; otherwise -1
)

# 3. Define the classification function
def classify_climate_change(example):
    text = example["text"][:1024]  # Truncate to first 1024 characters
    result = classifier(
        sequences=text,
        candidate_labels=["climate change"],
        multi_label=False
    )
    example["climate_change_score"] = result["scores"][0]
    return example

# 4. Apply to the train split (be mindful of dataset size)
ds_train_with_scores = ds["train"].map(
    classify_climate_change,
    batched=False,
    num_proc=4
)

# 5. Filter for articles with score >= 0.5
threshold = 0.5
ds_climate = ds_train_with_scores.filter(
    lambda ex: ex["climate_change_score"] >= threshold
)

print(f"Number of articles about climate change: {len(ds_climate)}")

# 6. Inspect a few
for i in range(min(5, len(ds_climate))):
    print(ds_climate[i]["title"], ds_climate[i]["climate_change_score"])
    print(ds_climate[i]["text"][:200], "...")
    print("-" * 40)
```
